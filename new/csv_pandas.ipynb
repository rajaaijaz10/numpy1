# Load dataset

```python
import numpy as np
import pandas as pd
tt=pd.read_csv(r'titanic_train.csv')
tt1=pd.read_csv(r'titanic_train.csv')

#tt.keys()
tt
```

# Fill nan values in age with average/mean

```python
tt['Age'].fillna(tt['Age'].mean(), inplace=True)
tt
```

# Fill nan values in Cabin with Bfill 

```python
tt['Cabin'].fillna(method='bfill',inplace= True)
tt
```

# Fill remaning nan values in Cabin with Ffill

```python
tt['Cabin'].fillna(method='ffill',inplace= True)
tt
```

# Count of nulls in each col using loop

```python
col=tt1.keys()
for ss in col:
    x=tt1[ss].isnull().sum()
    if x<1:
        continue
    print ("%s ---- %d" %(ss,x) )
```
